```python
import pandas as pd

# Function to extract the numeric value after "cid-"
def extract_cid_value(text):
    start_index = text.find('cid-') + 4
    end_index = text.find('-', start_index)
    return text[start_index:end_index]

df['CID'] = df['id'].apply(extract_cid_value)
```

```python
valuesets_counts = df.groupby('display').agg({'code': 'nunique', 'system': 'nunique', 'CID': 'nunique'}).reset_index()
valuesets_counts.columns = ['display', 'ucount_code', 'ucount_system', 'ucount_CID']

valuesets_counts[(valuesets_counts['ucount_code']>1) | (valuesets_counts['ucount_system']>1)]
```

```python
df[df['display'] == 'Angle']
```

```python
multi_use_valuesets = valuesets_counts[(valuesets_counts['ucount_code']>1) | (valuesets_counts['ucount_system']>1)]
pd.merge(df, multi_use_valuesets, on = 'display').groupby('system')['display'].nunique().sort_values(ascending=False)
```

```python
pd.merge(df, multi_use_valuesets, on = 'display').pivot_table(index='display', columns='system', aggfunc='size', fill_value=0)
```
